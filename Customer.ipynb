```python
import sklearn
import pandas as pd
df=pd.read_csv("insurance-ltv-decrease.csv")
df
```

```python
df.loc[13933].income
cond=df['income'] == '<=2L'
df=df[cond==False]
df.income.value_counts()
```

```python
df['income']=df['income'].map({'2L-5L':'2L-5L','5L-10L':'5L-10L','More than 10L': '10L-25L'})
df.income.unique()
```

```python
df.columns
```

```python
df[['lower_income','higher_income']]=df['income'].str.split('-',n=1,expand=True)
df['higher_income']=df['higher_income'].str.rstrip('L')
df['lower_income']=df['lower_income'].str.rstrip('L')

df.sample()
```

```python
df=df.drop('income',axis=1)
df.columns
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.higher_income.unique()
df.lower_income.unique()
```

```python
df
```

```python
df['gender']=df['gender'].map({'Male':1,'Female':0})
df['area']=df['area'].map({'Rural':0,'Urban':1})
df['qualification']=df['qualification'].map({'Bachelor':0,'High School':1,'Others':2})
df['num_policies']=df['num_policies'].map({'More than 1':0,'1':1})
df['policy']=df['policy'].map({'A':0,'B':1,'C':2})
df['type_of_policy']=df['type_of_policy'].map({'Platinum':2,'Gold':1,'Silver':0})
df['higher_income']=df['higher_income'].map({'10':10,'5':5,'25':25})
df['lower_income']=df['lower_income'].map({'5':5,'2':2,'10':10})

df.sample(10)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

model=LinearRegression()
t=df['cltv']
feat=df[['gender', 'area', 'qualification','marital_status','claim_amount','num_policies', 'type_of_policy','higher_income','lower_income']]

X_train, X_test, y_train, y_test = train_test_split(feat, t, test_size=0.150, random_state=30)

model.fit(X_train,y_train)
s=model.predict(X_test)

r2 = r2_score(y_test,s)
print("The r2 score is :",r2)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestRegressor

grid = {
    'n_estimators': [100, 200, 300],
    'max_depth': [None, 10, 20],
    'min_samples_split': [2, 5, 10]
}

search = GridSearchCV(RandomForestRegressor(), grid, cv=5, scoring='r2')

search.fit(X_train, y_train)

para = search.para_
print("Best parameters:", para)

best = RandomForestRegressor(**para)

best.fit(X_train, y_train)

y_pred = best.predict(X_test)

r2 = r2_score(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
rmse = mean_squared_error(y_test, y_pred, squared=False)

print(f"R-squared: {r2}")
print(f"Mean Absolute Error (MAE): {mae}")
print(f"Mean Squared Error (MSE): {mse}")
print(f"Root Mean Squared Error (RMSE): {rmse}")
```
